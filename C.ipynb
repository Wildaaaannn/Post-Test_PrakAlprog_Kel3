```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Konfigurasi visualisasi
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 7))

# Membaca data
df = pd.read_csv('Kelas A_Gender Inequality Index.csv')
df_clean = df.dropna(subset=['M_secondary_educ', 'F_secondary_educ', 'GII', 'Human_development']).copy()

# Menghitung selisih edukasi (Education Gap)
df_clean['Educ_Gap'] = df_clean['M_secondary_educ'] - df_clean['F_secondary_educ']

# Membuat garis regresi (trendline)
sns.regplot(
    x='Educ_Gap', 
    y='GII', 
    data=df_clean, 
    scatter=False, 
    color='black', 
    line_kws={'linestyle':'--', 'linewidth': 2} 
)

# Membuat scatter plot di atas garis regresi
sns.scatterplot(
    x='Educ_Gap', 
    y='GII', 
    hue='Human_development', 
    palette='Set1',
    data=df_clean, 
    s=100, 
    alpha=0.8,
    edgecolor='black'
)

# Kustomisasi label sumbu
plt.title('Hubungan Education Gap dengan Gender Inequality Index (GII)', 
          fontsize=15, fontweight='bold', pad=15)
plt.xlabel('Education Gap (M_secondary_educ - F_secondary_educ)', fontsize=12, fontweight='bold')
plt.ylabel('Gender Inequality Index (GII)', fontsize=12, fontweight='bold')

# Memindahkan legenda agar tidak menutupi grafik
plt.legend(title='Tingkat Development', bbox_to_anchor=(1.02, 1), loc='upper left')

plt.tight_layout()
# Menyimpan grafik sesuai penamaan soal
plt.savefig('Soal_3_Scatter_Plot_Education_Gap.png', dpi=300, bbox_inches='tight')
```
