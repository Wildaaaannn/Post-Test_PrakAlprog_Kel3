```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Konfigurasi visualisasi modern
sns.set_theme(style="white", rc={"axes.facecolor": "#ffffff"})
plt.figure(figsize=(10, 6))

# Membaca data dan membuang nilai kosong
df = pd.read_csv('Kelas A_Gender Inequality Index.csv')
df_clean = df.dropna(subset=['Human_development', 'Maternal_mortality'])

# Agregasi standar deviasi dan pengurutan data
std_data = df_clean.groupby('Human_development')['Maternal_mortality'].std().reset_index()
std_data = std_data.sort_values('Maternal_mortality', ascending=False)

# Membuat bar chart
ax = sns.barplot(
    x='Human_development', 
    y='Maternal_mortality', 
    data=std_data, 
    palette='mako', 
    edgecolor='black',
    linewidth=1.2,
    width=0.6 
)

# Menyesuaikan elemen estetika grafik
sns.despine()
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Kustomisasi label dan judul
plt.title('Standar Deviasi Maternal Mortality\nBerdasarkan Tingkat Human Development', 
          fontsize=15, fontweight='bold', pad=20)
plt.xlabel('Tingkat Human Development', fontsize=12, fontweight='bold', labelpad=10)
plt.ylabel('Standar Deviasi Maternal Mortality', fontsize=12, fontweight='bold', labelpad=10)

# Menambahkan anotasi nilai pada bar
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:.2f}', 
                (p.get_x() + p.get_width() / 2., height), 
                ha='center', va='bottom', 
                xytext=(0, 8), textcoords='offset points',
                fontsize=11, fontweight='bold')

plt.tight_layout()
# Menyimpan grafik sesuai penamaan soal
plt.savefig('Soal_1_Standar_Deviasi_Maternal_Mortality.png', dpi=300, bbox_inches='tight')
```
